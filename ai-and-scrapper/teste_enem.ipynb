```python
import fitz  # PyMuPDF
import os
import re
import json

# --- Configurações iniciais ---
PDF_PATH = "prova_enem.pdf"
IMAGEM_DIR = "imagens"
JSON_DIR = "jsons"

os.makedirs(IMAGEM_DIR, exist_ok=True)
os.makedirs(JSON_DIR, exist_ok=True)

# --- Função para extrair imagens de uma página ---
def extrair_imagens(pagina, page_number):
    imagens_extraidas = []
    for i, img in enumerate(pagina.get_images(full=True)):
        xref = img[0]
        base_img = pagina.parent.extract_image(xref)
        image_bytes = base_img["image"]
        image_ext = base_img["ext"]
        nome_arquivo = f"img_{page_number}_{i}.{image_ext}"
        caminho = os.path.join(IMAGEM_DIR, nome_arquivo)
        with open(caminho, "wb") as f:
            f.write(image_bytes)
        imagens_extraidas.append(caminho)
    return imagens_extraidas

# --- Expressão regular para capturar questões ---
regex_questao = re.compile(
    r"QUESTÃO\s+(?P<num>\d{2,})\s*(?P<enunciado>.*?)(?=\n[A-E](?!\w))",
    re.DOTALL
)


regex_alternativas = re.compile(
    r"\n([A-E])\s+(.*?)(?=\n[A-E]\s+|\nQUESTÃO|\Z)",
    re.DOTALL
)


# --- Abrir PDF ---
doc = fitz.open(PDF_PATH)
questao_id = 1

for page_number, pagina in enumerate(doc):
    texto = pagina.get_text()
    imagens = extrair_imagens(pagina, page_number)

    # Extrair todas as questões da página
    questoes = list(regex_questao.finditer(texto))

    for questao in questoes:
        num = questao.group("num")
        enunciado_raw = questao.group("enunciado").strip()

        # Pega o texto a partir do enunciado até o final da página (para achar alternativas)
        idx_inicio = texto.find(questao.group(0))
        trecho = texto[idx_inicio:]

        # Extrair alternativas
        alternativas = {}
        for alt in regex_alternativas.finditer(trecho):
            letra = alt.group(1).upper()
            texto_alt = alt.group(2).strip()
            alternativas[letra] = texto_alt

        # Criar estrutura JSON
        questao_json = {
            "id": f"enem_q{questao_id}",
            "numero": int(num),
            "pagina": page_number + 1,
            "enunciado": enunciado_raw,
            "alternativas": alternativas,
            "imagem_urls": imagens  # pode conter várias
        }

        # Salvar JSON
        arquivo_json = os.path.join(JSON_DIR, f"q{questao_id}.json")
        with open(arquivo_json, "w", encoding="utf-8") as f:
            json.dump(questao_json, f, ensure_ascii=False, indent=2)

        print(f"Questão {questao_id} salva.")
        questao_id += 1

print("✅ Extração finalizada.")
```

```python
import os
import re
import json
from pdf2image import convert_from_path
from PIL import Image
import pytesseract

# --- Configuração ---
PDF_PATH = "prova_enem.pdf"
IMAGEM_DIR = "imagens"
JSON_DIR = "jsons"

os.makedirs(IMAGEM_DIR, exist_ok=True)
os.makedirs(JSON_DIR, exist_ok=True)

# --- Etapa 1: Converter PDF em imagens ---
print("📄 Convertendo PDF em imagens...")
paginas = convert_from_path(PDF_PATH, dpi=300)

# --- Etapa 2: Processar cada imagem com OCR ---
regex_questao = re.compile(r"QUESTÃO\s+(\d{2,})", re.IGNORECASE)
regex_alternativa = re.compile(r"^[A-E]\s+.*", re.IGNORECASE)

questao_id = 0
for i, img in enumerate(paginas):
    img_path = os.path.join(IMAGEM_DIR, f"pagina_{i+1}.png")
    img.save(img_path)

    print(f"🔍 OCR página {i+1}...")
    texto = pytesseract.image_to_string(Image.open(img_path), lang="por")

    # --- Etapa 3: Separar por questões ---
    blocos = re.split(r"\bQUESTÃO\s+\d{2,}", texto)
    numeros = regex_questao.findall(texto)

    for j, bloco in enumerate(blocos[1:]):  # ignorar cabeçalho antes da primeira "QUESTÃO"
        num = numeros[j]
        linhas = bloco.strip().split("\n")

        enunciado = []
        alternativas = {}
        letra_atual = None

        for linha in linhas:
            linha = linha.strip()
            if not linha:
                continue

            if re.match(r"^[A-E]\s+", linha):
                letra_atual = linha[0].upper()
                texto_alt = linha[2:].strip()
                alternativas[letra_atual] = texto_alt
            elif letra_atual:
                alternativas[letra_atual] += " " + linha
            else:
                enunciado.append(linha)

        dados = {
            "id": f"enem_q{questao_id + 1}",
            "numero": int(num),
            "pagina": i + 1,
            "enunciado": " ".join(enunciado),
            "alternativas": alternativas,
            "imagem_pagina": img_path
        }

        with open(os.path.join(JSON_DIR, f"q{questao_id + 1}.json"), "w", encoding="utf-8") as f:
            json.dump(dados, f, ensure_ascii=False, indent=2)

        print(f"✅ Questão {questao_id + 1} extraída.")
        questao_id += 1

print("🏁 Extração finalizada.")
```
